```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("SOCR_Data_Dinov_020108_HeightsWeights.txt", delim_whitespace=True )
df.head(5)
```

```python
height = np.array(df["Height(Inches)"])
height = height.reshape(-1,1)
weight = np.array(df["Weight(Pounds)"])
weight = weight.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(height, weight)
```

```python
reg.predict(68.87761 )
```

```python
plt.figure(figsize=(20, 10))
plt.xlabel("Heights(Inches)", fontsize=20)
plt.ylabel("Weight(Pounds)", fontsize=20)
plt.scatter(height, weight, edgecolors='k', facecolors='none')
plt.plot(height, reg.predict(height), color='r')
```
